# **5.** **HYPOTHESIS TESTING:**

**Hypothesis:**
Running pace is slower on days with poor air quality (high AQI values) compared to days with good air quality.

**Null Hypothesis (H):**
AQI has no significant effect on my pace.

**Alternative Hypothesis (H):**
The mean running pace is slower on unhealthy air quality days compared to good air quality days.



```python
from scipy.stats import ttest_ind

# Choosing the "Good" ve "Unhealthy" AQI categories:
good_pace = df[df["AQI_Category"] == "Good"]["Pace (min/km)"].dropna()
unhealthy_pace = df[df["AQI_Category"] == "Unhealthy"]["Pace (min/km)"].dropna()

# One-way T-Test:
t_stat, p_value = ttest_ind(unhealthy_pace, good_pace, alternative='greater')

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference between pace and AQI.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between pace and AQI.")
```

So, the T-test suggest that **the null hypothesis should be rejected.**

```python
# ANOVA:
from scipy.stats import f_oneway
df["AQI_Category"] = df["AQI"].apply(categorize_aqi)

# Creating a list of pace values grouped by AQI category:
groups = [df[df["AQI_Category"] == cat]["Pace (min/km)"].dropna()
          for cat in ["Good", "Moderate", "Unhealthy for Sensitive Groups", "Unhealthy"]]

# Running one-way ANOVA:
f_stat, p_value = f_oneway(*groups)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference between pace and AQI.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between pace and AQI.")
```

So, the ANOVA test suggest that **the null hypothesis should be rejected.**

# **6.** **INSIGHTS:**

* **Running pace tends to be slower on days with worse air quality.** Especially on days when AQI was high, my average pace increased, which suggests I might be running more slowly in polluted conditions. This supports my original hypothesis.

* **Heart rate also seems to go up slightly when AQI, PM2.5, or temperature increases.** This could mean that my body is working harder when the air isnt as clean or when its hotter outside.

* Running distance doesnt change as dramatically, but **on some high-AQI or very hot days, I ran shorter distances**. This might be because I stopped earlier due to breathing discomfort or lower energy.

* Both the **T-test and ANOVA test** showed that there is a statistically significant difference in my pace between different AQI categories, which means the variation I saw in the plots is not just by chance.

# **7.** **MACHINE LEARNING METHODS:**

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score
from sklearn.feature_selection import f_regression, RFECV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from matplotlib.ticker import ScalarFormatter

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

link = "https://raw.githubusercontent.com/yagmurtomruk/yagmurtomruk/main/DSA210_DATA.xlsx"
df = pd.read_excel(link)

# Defining the feature and target lists
all_features = ['Temperature', 'Humidity', 'AQI', 'PM2.5 conc.', 'NO2 conc']
targets = ['Pace (min/km)', 'Heart Rate (BPM)', 'Running Distance (km)']

# Defining model set from lectures
model_defs = {
    'LinearRegression': (Pipeline([('scaler', StandardScaler()), ('reg', LinearRegression())]), {}),
    'Ridge': (Pipeline([('scaler', StandardScaler()), ('reg', Ridge())]), {'reg__alpha': [0.1, 1, 10]}),
    'Lasso': (Pipeline([('scaler', StandardScaler()), ('reg', Lasso(max_iter=5000))]), {'reg__alpha': [0.01, 0.1, 1]}),
    'RandomForest': (RandomForestRegressor(random_state=42), {'n_estimators': [100, 200], 'max_depth': [None, 5, 10]}),
    'GradientBoosting': (GradientBoostingRegressor(random_state=42), {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 5]}),
    'SVR': (Pipeline([('scaler', StandardScaler()), ('reg', SVR())]), {'reg__C': [0.1, 1, 10], 'reg__gamma': ['scale', 'auto']}),
    'AdaBoost': (AdaBoostRegressor(random_state=42), {'n_estimators': [50, 100], 'learning_rate': [0.01, 0.1, 1]})
}

# CV
imputer = SimpleImputer(strategy='median')
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Creating a loop for modelling:

for target in targets:

    df = pd.read_excel(link)

    feature_cols = [f for f in all_features if f != target]
    data = df[feature_cols + [target]].dropna(subset=[target]).reset_index(drop=True)
    X_raw = data[feature_cols]
    y = data[target]

    X_imp = pd.DataFrame(imputer.fit_transform(X_raw), columns=feature_cols)

    # F-score ranking
    f_scores, _ = f_regression(X_imp, y)
    scored = pd.Series(f_scores, index=feature_cols).sort_values(ascending=False)
    ranked = scored

    subsets = {
        'Top5': ranked.index[:5].tolist(),
        'Top2': ranked.index[:2].tolist()
    }

    rfecv = RFECV(RandomForestRegressor(random_state=42), step=1, cv=cv,
                  scoring='neg_mean_squared_error', n_jobs=-1).fit(X_imp, y)
    subsets['RFE'] = list(X_imp.columns[rfecv.support_])

    # Feature importance plots
    for subset_name, feat_list in subsets.items():
        vals = scored[feat_list]
        fig, ax = plt.subplots(figsize=(5, 3))
        bars = ax.bar(vals.index, vals.values)
        ax.set_title(f'{target} - {subset_name} Feature F-scores')
        ax.set_ylabel('F-score')
        ax.set_xticklabels(vals.index, rotation=45)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        for bar in bars:
            h = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2, h, f'{h:.2f}', ha='center', va='bottom')
        fig.tight_layout()
        plt.show()

    # Regression model results
    n = len(subsets)
    fig, axes = plt.subplots(n, 2, figsize=(12, 5 * n))

    for i, (subset_name, feat_list) in enumerate(subsets.items()):
        X_sub = X_imp[feat_list]
        mses, maes, r2s, names = [], [], [], []

        for model_name, (model, params) in model_defs.items():
            gs = GridSearchCV(model, params, cv=cv, scoring='neg_mean_squared_error', n_jobs=-1)
            gs.fit(X_sub, y)
            best = gs.best_estimator_
            preds = best.predict(X_sub)

            mse = -gs.best_score_
            mae = mean_absolute_error(y, preds)
            r2 = r2_score(y, preds)

            mses.append(mse)
            maes.append(mae)
            r2s.append(r2)
            names.append(model_name)

            axes[i, 0].scatter(y, preds, alpha=0.6, label=f'{model_name} (R={r2:.2f})')

        axes[i, 0].spines['top'].set_visible(False)
        axes[i, 0].spines['right'].set_visible(False)
        axes[i, 0].plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
        axes[i, 0].set_title(f'{target} - {subset_name} Predictions')
        axes[i, 0].set_xlabel('Actual')
        axes[i, 0].set_ylabel('Predicted')
        axes[i, 0].legend(fontsize=8)

        x = np.arange(len(names))
        w = 0.35
        bars1 = axes[i, 1].bar(x - w/2, mses, w, label='MSE')
        bars2 = axes[i, 1].bar(x + w/2, maes, w, label='MAE')
        axes[i, 1].set_xticks(x)
        axes[i, 1].set_xticklabels(names, rotation=45)
        axes[i, 1].set_title(f'{target} - {subset_name} Errors')
        axes[i, 1].set_ylabel('Error')
        axes[i, 1].yaxis.set_major_formatter(ScalarFormatter())
        axes[i, 1].ticklabel_format(axis='y', style='plain', useOffset=False)
        axes[i, 1].spines['top'].set_visible(False)
        axes[i, 1].spines['right'].set_visible(False)
        axes[i, 1].legend()

        for bar in bars1:
            h = bar.get_height()
            axes[i, 1].text(bar.get_x() + bar.get_width()/2, h, f'{h:.2f}', ha='center', va='bottom')
        for bar in bars2:
            h = bar.get_height()
            axes[i, 1].text(bar.get_x() + bar.get_width()/2, h, f'{h:.2f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```
